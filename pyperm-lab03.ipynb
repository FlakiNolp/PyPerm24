### Немного теории для задачи 6*

В Python есть функция `eval()`, которая позволяет оценивать выражение, написанное внутри строки как код. 

**Пример 1.** Перечень элементов в квадратных скобках внутри строки превращается в список:

    eval("[3, 7, 8]")
    [3, 7, 8]
    
**Пример 2.** Перечень элементов в фигурных скобках внутри строки превращается в словарь:

    eval("{'A' : 3, 'B' : 9}")
    {'A': 3, 'B': 9}


**Пример 3.** Выражение в строке оценивается как математическое выражение, если `x` определён:
    
    x = 10
    eval("3 * x")
    30
    
**Пример 4.** Выражение в строке оценивается как полноценный питоновский код с функцией:

    eval("sum([3, 7, 5])")
    15
    
    
Функция `eval()` является довольно мощной, но при этом опасной, так как она не проверяет код, который написан внутри строки. То есть, если в примере 4 вместо безобидной функции `sum()` был бы написан код, который стирает все файлы на компьютере или что-то подобное, он всё равно был бы запущен при применении функции `eval()`. Поэтому эту функцию рекомендуют использовать с осторожностью или вообще заменять её на аналог `literal_eval()` из модуля `ast` (правда, `literal_eval()` не умеет работать с выражениями, содержащими операторы и функции, из примеров выше она сработала бы только в первых двух).
